```python
# STEP 1: Import the necessary modules.

#mediapipe 0.10.35, most recent one compatable with any python version that's self respecting (since 2023)
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision

# STEP 2: Create an HandLandmarker object.

# Install hand_landmarker.task from https://ai.google.dev/edge/mediapipe/solutions/vision/hand_landmarker/index#models
# Under Models click the link for hand marker, put that file in the same directory as this ipynb notebook
base_options = python.BaseOptions(model_asset_path='hand_landmarker.task')
options = vision.HandLandmarkerOptions(
    base_options=base_options,
    
    # equivalent to static_image_mode=True
    running_mode=vision.RunningMode.IMAGE,

    # equivalent to max_num_hands=1
    num_hands=1,

    # equivalent to min_detection_confidence=0.5
    min_hand_detection_confidence=0.5
)
detector = vision.HandLandmarker.create_from_options(options)
```

```python
import cv2
import numpy as np

# MediaPipe hand connections (same as official spec)
HAND_CONNECTIONS = [
    (0, 1), (1, 2), (2, 3), (3, 4),
    (0, 5), (5, 6), (6, 7), (7, 8),
    (5, 9), (9, 10), (10, 11), (11, 12),
    (9, 13), (13, 14), (14, 15), (15, 16),
    (13, 17), (0, 17),
    (17, 18), (18, 19), (19, 20)
]


def draw_landmarks_on_image(rgb_image, detection_result):
    image = np.copy(rgb_image)
    h, w, _ = image.shape

    for hand_landmarks in detection_result.hand_landmarks:
        # Convert normalized coords -> pixel coords
        points = []
        for lm in hand_landmarks:
            x, y = int(lm.x * w), int(lm.y * h)
            points.append((x, y))

        # Draw points
        for x, y in points:
            cv2.circle(image, (x, y), 5, (0, 255, 0), -1)

        # Draw connections
        for a, b in HAND_CONNECTIONS:
            if a < len(points) and b < len(points):
                cv2.line(image, points[a], points[b], (255, 0, 0), 2)

    return image
```

```python
from tqdm import tqdm

SEQUENCE_LENGTH = 37
NUM_LANDMARKS = 21
FEATURES_PER_FRAME = NUM_LANDMARKS * 3  # x, y, z = 63

# Function Nghi made

# Image path - > img input
# update mediapipe
# assumes image_array is rgb
def extract_landmarks_from_image(image_array):

    # Image handling

    if image_array is None:
        return np.zeros(FEATURES_PER_FRAME)

    # Mediapipe Stuff
    mp_image = mp.Image(
        image_format=mp.ImageFormat.SRGB,
        data=image_array
    )
    
    detection_result = detector.detect(mp_image)

    #drawn_result = draw_landmarks_on_image(image_array, detection_result)

    #plt.imshow(drawn_result)
    #plt.show()

    if detection_result.hand_landmarks:
        hand_landmarks = detection_result.hand_landmarks[0]

        landmarks = []
        for lm in hand_landmarks:
            landmarks.extend([lm.x, lm.y, lm.z])

        return np.array(landmarks)

    else:
        return np.zeros(FEATURES_PER_FRAME)
```

```python
# TODO Implement model which takes in (37, 63) numpy array
# outputs index of class

# NOTE: Ensure installed packages/imports do not conflict within the current conda environment.
# =========================================================
# IMPORTS
# =========================================================

import numpy as np
import torch
import torch.nn as nn


# =========================================================
# TRANSFORMER MODEL FOR GESTURE CLASSIFICATION
#
# INPUT:
#   (batch_size, 37 timesteps, 63 features)
#
# OUTPUT:
#   logits for 27 gesture classes
#
# Example input shape:
#   (8, 37, 63)
#
# Example output shape:
#   (8, 27)
# =========================================================

class GestureTransformer(nn.Module):

    def __init__(
        self,
        input_dim=63,      # number of features per frame
        model_dim=128,     # transformer internal size
        num_heads=4,       # number of attention heads
        num_layers=3,      # number of transformer layers
        num_classes=27     # number of gesture classes
    ):

        # Initialize parent nn.Module class
        super().__init__()

        # =================================================
        # STEP 1:
        # Convert each frame from 63 features
        # into a larger vector (128 features)
        #
        # Shape:
        # (batch, 37, 63)
        # ->
        # (batch, 37, 128)
        # =================================================
        self.input_projection = nn.Linear(
            input_dim,
            model_dim
        )

        # =================================================
        # STEP 2:
        # Positional embeddings
        #
        # Transformers do NOT know order naturally.
        # We add position information manually.
        #
        # Shape:
        # (1, 37, 128)
        #
        # 37 = number of timesteps
        # =================================================
        self.pos_embedding = nn.Parameter(
            torch.randn(1, 37, model_dim)
        )

        # =================================================
        # STEP 3:
        # Create ONE transformer encoder layer
        #
        # d_model:
        #   size of feature vectors
        #
        # nhead:
        #   number of attention heads
        #
        # batch_first=True:
        #   input shape is
        #   (batch, sequence, features)
        # =================================================
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=model_dim,
            nhead=num_heads,
            batch_first=True
        )

        # =================================================
        # STEP 4:
        # Stack multiple transformer layers
        # =================================================
        self.transformer = nn.TransformerEncoder(
            encoder_layer,
            num_layers=num_layers
        )

        # =================================================
        # STEP 5:
        # Final classifier
        #
        # Takes transformer output
        # and predicts gesture class
        # =================================================
        self.classifier = nn.Sequential(

            # Reduce size
            nn.Linear(model_dim, 64),

            # Non-linearity
            nn.ReLU(),

            # Prevent overfitting
            nn.Dropout(0.2),

            # Final output layer
            nn.Linear(64, num_classes)
        )

    # =====================================================
    # FORWARD PASS
    #
    # Defines how data moves through model
    # =====================================================
    def forward(self, x):

        # =================================================
        # x shape:
        # (batch_size, 37, 63)
        # =================================================

        # =================================================
        # STEP 1:
        # Project 63 features -> 128 features
        # =================================================
        x = self.input_projection(x)

        # =================================================
        # STEP 2:
        # Add positional embeddings
        #
        # This helps transformer know frame order
        # =================================================
        x = x + self.pos_embedding

        # =================================================
        # STEP 3:
        # Pass through transformer
        #
        # Transformer learns relationships between
        # frames over time
        # =================================================
        x = self.transformer(x)

        # =================================================
        # STEP 4:
        # Average across all timesteps
        #
        # Before:
        # (batch, 37, 128)
        #
        # After:
        # (batch, 128)
        # =================================================
        x = x.mean(dim=1)

        # =================================================
        # STEP 5:
        # Predict gesture classes
        #
        # Output:
        # (batch, 27)
        # =================================================
        x = self.classifier(x)

        return x


# =========================================================
# TEST THE MODEL
# =========================================================

# Create model
model = GestureTransformer()

# Print model architecture
print(model)

# =========================================================
# CREATE DUMMY INPUT
#
# batch size = 8
# 37 frames
# 63 features per frame
# =========================================================
dummy_input = torch.randn(8, 37, 63)

print("\nInput shape:")
print(dummy_input.shape)

# =========================================================
# RUN FORWARD PASS
# =========================================================
output = model(dummy_input)

print("\nOutput shape:")
print(output.shape)

# Expected:
# torch.Size([8, 27])

# =========================================================
# GET PREDICTED CLASS
#
# argmax finds highest scoring class
# =========================================================
predictions = torch.argmax(output, dim=1)

print("\nPredicted classes:")
print(predictions)
```
